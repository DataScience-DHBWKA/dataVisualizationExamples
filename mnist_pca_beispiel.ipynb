```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA

# MNIST-Daten laden
mnist = fetch_openml('mnist_784')
X, y = mnist.data, mnist.target

# Filtere nur die Bilder der Ziffer 1
X_0 = X[y == '0']
X_1 = X[y == '1']
X_2 = X[y == '2']
X_2 = X[y == '2']
```

```python
# PCA durchführen
pca = PCA(n_components=10)
X_0_transformed = pca.fit_transform(X[y == '0'])
X_1_transformed = pca.fit_transform(X[y == '1'])
X_2_transformed = pca.fit_transform(X[y == '2'])
X_3_transformed = pca.fit_transform(X[y == '3'])
eigenfaces = pca.components_.reshape((10, 28, 28))
```

```python
# Eigenfaces anzeigen
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flatten()):
    ax.imshow(eigenfaces[i], cmap='gray')
    ax.set_title(f'Eigenface {i+1}')
    ax.axis('off')
plt.show()
```
